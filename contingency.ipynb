```python
import pandas as pd
import numpy as np
```

```python
df_1 = pd.read_csv('PG_campaigning.csv')
df_1
```

```python
df_1['Why did you vote for him?.1'].unique()
```

```python
filtered_df = df_1[df_1['Your Program'].isin(['UG', 'PG'])]
filtered_df
```

```python
contingency_table = pd.crosstab(filtered_df['Your Program'], filtered_df['Why did you vote for him?.1'], margins=True, margins_name='Total')
(contingency_table)
```

```python
import matplotlib.pyplot as plt

# Drop the 'Total' row to avoid duplicate bars
# contingency_table = contingency_table

# Plot the segmented bar graph
contingency_table.T.plot(kind='bar', stacked=True, figsize=(10, 6))

# Add labels and title
plt.xlabel('Program')
plt.ylabel('Count')
plt.title('Voting Reasons by Program')
plt.style('ggplot')

# Show the plot
plt.legend(title='Voting Reason')
plt.grid(True)
plt.show()
```
